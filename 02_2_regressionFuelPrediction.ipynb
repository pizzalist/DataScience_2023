```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/ralbu85/DataScience_2022S/master/data/auto.csv')
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.pairplot(df)
```

```python
df.isnull().sum()
```

```python
df = pd.get_dummies(df,columns=['origin'])
df
```

```python
y = df['mpg']
```

```python
X = df.drop(columns=['mpg'])
```

### 회귀분석 모듈 불러오기

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

### 학습

```python
reg.fit(X,y)
```

```python
y_pred = reg.predict(X)
y_pred[:5]
```

### 학습된 계수 구하기

```python
reg.coef_
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,8))
plt.bar(X.columns, reg.coef_,)
```

### 학습된 값과 정답간의 차이

```python
plt.scatter(y,y_pred)
```

```python
import numpy as np
np.sqrt(np.sum((y-y_pred)**2))
```

### 데이터를 정규화하여 학습

```python
X_std = (X-X.mean())/X.std()
X_std
```

```python
reg.fit(X_std,y)
```

```python
plt.figure(figsize=(12,8))
plt.bar(X_std.columns, reg.coef_)
```

```python
y_pred = reg.predict(X_std)
```

```python
plt.scatter(y,y_pred)
```

```python
np.sqrt(np.sum((y-y_pred)**2))
```
